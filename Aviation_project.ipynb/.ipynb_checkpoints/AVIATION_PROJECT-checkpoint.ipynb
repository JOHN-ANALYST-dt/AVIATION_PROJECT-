<h3>Lets begin by Filling the gaps </h3>

```python
#print out the datatypes to see which fomulae to use in cleaning 
Aviation.info()
```

<h3>We will Begin with filling the object datatypes, string with Mode</h3>

```python
#using the for loop we will fill all the string datatypes with mode 
for col in Aviation.select_dtypes(include='object').columns:
    if Aviation[col].isnull().any():
        mode_val = Aviation[col].mode()[0]
        Aviation[col].fillna(mode_val, inplace=True)
```

<h3>Fill all the intergers or float with meadian</h3>

<h4>Median is the best choice because it is likely to influence the distribution of the data set overall<br>
    <br>Why This Approach is Effective for Float Data:<br>
    <br>
Preserves Distribution<br>
    <br>:
Median maintains the central tendency without being affected by extreme value<br>s.

Handles Skewed D
<br>
<br>ata:
For income, age, or sensor readings where distributions are often skewed, median is more representative than <br>m<br>ean.

Computationally Eff
<br>icient:
Much faster than complex imputation methods</h4>

```python
#filling in the float with meadian
for col in Aviation.select_dtypes(include='float64').columns:
    if Aviation[col].isnull().any():
        median_val = Aviation[col].median()
        Aviation[col].fillna(median_val, inplace=True)
```

<h3>Plotting the Aircraft Make with the highest and Lowest incidence of Accidence</h3>

```python
Count_numb_Accident=Aviation["Make"].value_counts().head()
print(Count_numb_Accident)
```

```python
# Step 1: Count the number of accidents per Make
make_counts = Aviation["Make"].value_counts().head(5)

# Step 2: Get the Make with the highest and lowest accident count
highest_make = make_counts.idxmax()
highest_count = make_counts.max()

lowest_make = make_counts[make_counts == make_counts.min()].index[0]
lowest_count = make_counts.min()

# Step 3: Prepare data for plotting
summary = pd.Series({highest_make: highest_count, lowest_make: lowest_count})

# Step 4: Plot
summary.plot(kind='bar', color=['skyblue', 'steelblue'], figsize=(8, 5))

plt.title('Aircraft Make with Highest and Lowest Number of Accidents')
plt.ylabel('Number of Accidents')
plt.xlabel('Aircraft Make')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<h3>Trend of Fatal Injuries Over Time(Per year)</h3>

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert Date column to datetime (if it's not already)
Aviation['Date_time'] = pd.to_datetime(Aviation['Date_time'])

# Group by year and calculate total fatal injuries
Aviation['Year'] = Aviation['Date_time'].dt.year
yearly_fatalities = Aviation.groupby('Year')['Total.Fatal.Injuries'].sum().reset_index()


# Convert Date column to datetime (if it's not already)
Aviation['Date_time'] = pd.to_datetime(Aviation['Date_time'])

# Group by year and calculate total fatal injuries
Aviation['Year'] = Aviation['Date_time'].dt.year
yearly_fatalities = Aviation.groupby('Year')['Total.Fatal.Injuries'].sum().reset_index()

# Plot
plt.figure(figsize=(8, 5))
sns.lineplot(data=yearly_fatalities, x='Year', y='Total.Fatal.Injuries', marker='o', color='red')

plt.title('Total Fatal Injuries per Year')
plt.xlabel('Year')
plt.ylabel('Fatal Injuries')
plt.grid(True)
plt.tight_layout()
plt.show()


```

<h3>We need to Show the kind of Aircraft Make in market from the years upto 2023.<br>You need to chech at the Best model trending in the market </h3>


```python

make_year_counts = Aviation.groupby(['Year', 'Make']).size().unstack(fill_value=0)

# Step 2: (Optional) Limit to top 5 makes
top_makes = Aviation['Make'].value_counts().head(5).index
make_year_counts = make_year_counts[top_makes]

# Step 3: Plot
make_year_counts.plot(kind='bar', stacked=True, figsize=(14, 6), colormap='Set2')

plt.title('Aircraft Make Reports Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Reports')
plt.legend(title='Make', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

<h3>Histograms of the Total Fatal injuries per year</h3>

```python
# Step 1: Ensure Date_time is datetime format
Aviation['Date_time'] = pd.to_datetime(Aviation['Date_time'], errors='coerce')

# Step 2: Extract year
Aviation['Year'] = Aviation['Date_time'].dt.year

# Step 3: Group by year and sum fatal injuries
injuries_by_year = Aviation.groupby('Year')['Total.Fatal.Injuries'].sum().reset_index()

# Step 4: Plot histogram
plt.figure(figsize=(10, 6))
sns.barplot(data=injuries_by_year, x='Year', y='Total.Fatal.Injuries', color='crimson')

plt.title("Total Fatal Injuries per Year")
plt.xlabel("Year")
plt.ylabel("Total Fatal Injuries")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
